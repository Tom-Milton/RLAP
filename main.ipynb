```python
from tkinter import *
from tkinter import ttk

from SAC_RLAP.run import run
from SAC_RLAP.slider_widget import Slider


def main():

    window = Tk()
    window.geometry("300x550")

    def disable_event(): pass
    window.protocol("WM_DELETE_WINDOW", disable_event)

    ttk.Separator(window, orient='vertical').grid(row=0, column=1, rowspan=26, sticky=NS)

    Label(window, text='Select Model').grid(row=0, column=0, sticky=NW)
    model = StringVar(window, 'SAC')
    Radiobutton(window, text='SAC', value='SAC', variable=model).grid(row=1, column=0, sticky=NW)
    Radiobutton(window, text='SAC_PER', value='SAC_PER', variable=model).grid(row=2, column=0, sticky=NW)
    Radiobutton(window, text='SAC_PER_ERE', value='SAC_PER_ERE', variable=model).grid(row=3, column=0, sticky=NW)
    Radiobutton(window, text='PID', value='PID', variable=model).grid(row=4, column=0, sticky=NW)

    Label(window, text='Select Processes').grid(row=0, column=2, sticky=NW)
    train = BooleanVar(window, True)
    play = BooleanVar(window, True)
    results = BooleanVar(window, True)
    print_results = BooleanVar(window, True)
    Checkbutton(window, text='train', onvalue=True, offvalue=False, variable=train).grid(row=1, column=2, sticky=NW)
    Checkbutton(window, text='play', onvalue=True, offvalue=False, variable=play).grid(row=2, column=2, sticky=NW)
    Checkbutton(window, text='results', onvalue=True, offvalue=False, variable=results).grid(row=3, column=2, sticky=NW)
    Checkbutton(window, text='print_results', onvalue=True, offvalue=False, variable=print_results).grid(row=4, column=2, sticky=NW)

    ttk.Separator(window, orient='horizontal').grid(row=5, column=0, columnspan=3, sticky=EW)

    Label(window, text='Select Meal Scenario').grid(row=6, column=0, sticky=NW)
    scenario = StringVar(window, 'harris_benedict')
    Radiobutton(window, text='no_meals', value='no_meals', variable=scenario).grid(row=7, column=0, sticky=NW)
    Radiobutton(window, text='random_scaled', value='random_scaled', variable=scenario).grid(row=8, column=0, sticky=NW)
    Radiobutton(window, text='harris_benedict', value='harris_benedict', variable=scenario).grid(row=9, column=0, sticky=NW)

    Label(window, text='Select Patients').grid(row=6, column=2, sticky=NW)
    adult = BooleanVar(window, True)
    adolescent = BooleanVar(window, True)
    child = BooleanVar(window, True)
    Checkbutton(window, text='adult', onvalue=True, offvalue=False, variable=adult).grid(row=7, column=2, sticky=NW)
    Checkbutton(window, text='adolescent', onvalue=True, offvalue=False, variable=adolescent).grid(row=8, column=2, sticky=NW)
    Checkbutton(window, text='child', onvalue=True, offvalue=False, variable=child).grid(row=9, column=2, sticky=NW)
    patient_selection = Slider(window, init_lis=[1, 5, 8, 11], min_val=1, max_val=11)
    patient_selection.grid(row=10, column=2, sticky=NW)

    ttk.Separator(window, orient='horizontal').grid(row=11, column=0, columnspan=3, sticky=EW)

    Label(window, text='Select Save Mode').grid(row=12, column=0, sticky=NW)
    save_mode = StringVar(window, 'validate')
    Radiobutton(window, text='every_time', value='every', variable=save_mode).grid(row=13, column=0, sticky=NW)
    Radiobutton(window, text='best_reward', value='best', variable=save_mode).grid(row=14, column=0, sticky=NW)
    Radiobutton(window, text='validation_score', value='validate', variable=save_mode).grid(row=15, column=0, sticky=NW)

    Label(window, text='Select State Space').grid(row=12, column=2, sticky=NW)
    weight = BooleanVar(window, True)
    age = BooleanVar(window, True)
    height = BooleanVar(window, True)
    tdi = BooleanVar(window, True)
    meals = BooleanVar(window, False)
    Checkbutton(window, text='weight', onvalue=True, offvalue=False, variable=weight).grid(row=13, column=2, sticky=NW)
    Checkbutton(window, text='age', onvalue=True, offvalue=False, variable=age).grid(row=14, column=2, sticky=NW)
    Checkbutton(window, text='height', onvalue=True, offvalue=False, variable=height).grid(row=15, column=2, sticky=NW)
    Checkbutton(window, text='tdi', onvalue=True, offvalue=False, variable=tdi).grid(row=16, column=2, sticky=NW)
    Checkbutton(window, text='meals', onvalue=True, offvalue=False, variable=meals).grid(row=17, column=2, sticky=NW)

    ttk.Separator(window, orient='horizontal').grid(row=18, column=0, columnspan=3, sticky=EW)

    Label(window, text='Select Timesteps').grid(row=19, column=0, sticky=NW)
    timesteps = StringVar(window, '250_000')
    Radiobutton(window, text='100_000', value='100_000', variable=timesteps).grid(row=20, column=0, sticky=NW)
    Radiobutton(window, text='250_000', value='250_000', variable=timesteps).grid(row=21, column=0, sticky=NW)
    Radiobutton(window, text='500_000', value='500_000', variable=timesteps).grid(row=22, column=0, sticky=NW)

    Label(window, text='Enter Load Run Name').grid(row=19, column=2, sticky=NW)
    load_name = StringVar(window, '')
    Entry(window, textvariable=load_name).grid(row=20, column=2, sticky=NW)

    Label(window, text='Enter Save Run Name').grid(row=21, column=2, sticky=NW)
    save_name = StringVar(window, '')
    Entry(window, textvariable=save_name).grid(row=22, column=2, sticky=NW)

    ttk.Separator(window, orient='horizontal').grid(row=23, column=0, columnspan=3, sticky=EW)
    
    Button(window, text='Run', command=window.destroy).grid(row=24, column=1, sticky=NW)

    mainloop()

    patient_selection = [int(i) for i in patient_selection.getValues()]
    patient_classes = [adult.get(), adolescent.get(), child.get()]
    state_space = [weight.get(), age.get(), height.get(), tdi.get(), meals.get()]
    processes = [train.get(), play.get(), results.get(), print_results.get()]
    timesteps = int(timesteps.get())

    run(model.get(), scenario.get(), save_mode.get(), timesteps, patient_classes, patient_selection, state_space, processes, save_name.get(), load_name.get())


if __name__ == '__main__': main()
```
